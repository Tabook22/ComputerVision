Line Graph Example
Let's say we have 5 points and we want to connect them with lines. Here's how to do it:

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Create a line plot
plt.plot(x, y)

# Add title and labels
plt.title("A Simple Line Graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show the plot
plt.show()
```

If you have two sets of data that you want to plot on the same graph, you can easily do that with matplotlib.pyplot. You just need to call plt.plot() twice before calling plt.show().

Here's an example where we plot two sets of data on a single line graph:

```python
import matplotlib.pyplot as plt

# Data set 1
x1 = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]

# Data set 2
x2 = [0, 1, 2, 3, 4]
y2 = [0, 1, 2, 3, 4]

# Create the line graph for the first set of data
plt.plot(x1, y1, label='Data set 1')

# Create the line graph for the second set of data
plt.plot(x2, y2, label='Data set 2')

# Add a title and axis labels
plt.title('Two Sets of Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend to differentiate the two sets of data
plt.legend()

# Show the graph
plt.show()
```

In this example:

The first plt.plot(x1, y1, label='Data set 1') plots the first set of data with x1 and y1. The label='Data set 1' gives it a label for the legend.</br>

The second plt.plot(x2, y2, label='Data set 2') plots the second set of data with x2 and y2. The label='Data set 2' gives it a label for the legend.</br>

plt.legend() adds a legend to the graph, which helps you tell which line corresponds to which data set.</br>

Run this code, and you'll see a graph with both sets of data plotted as lines. The legend will help you identify which line corresponds to which set of data.</br>

**Bar Chart**
In a bar chart, you typically have categories on the x-axis and their values on the y-axis. Here's how you can make a bar chart:

```python
import matplotlib.pyplot as plt

# Data
x = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry']
y = [5, 3, 9, 2, 8]

# Create a bar chart
plt.bar(x, y)

# Add title and labels
plt.title("Fruit Count")
plt.xlabel("Fruits")
plt.ylabel("Count")

# Show the plot
plt.show()
```

**Pie Chart**
In a pie chart, you typically show the proportion of different categories. Here's how you can make a pie chart:

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry']
sizes = [5, 3, 9, 2, 8]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add title
plt.title("Fruit Count")

# Show the plot
plt.show()
```
