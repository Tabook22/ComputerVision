This script will:

Allow you to upload an image file from your local computer.
Convert the uploaded image to grayscale.
Transform the grayscale image into a NumPy array (matrix of pixels).
Print the matrix and display the grayscale image using matplotlib (optional).
Just run this code in your Google Colab notebook, and you should be able to upload an image and see its grayscale pixel matrix.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload files
uploaded = files.upload()
```

```python
# Ensure only one image is uploaded
if len(uploaded) == 1:
    filename = next(iter(uploaded))
    # Read the image using OpenCV
    original_image = cv2.imread(filename)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
    # Resize the image to 12x12 pixels
    resized_image = cv2.resize(gray_image, (30, 30))

    # Convert the original BGR image to RGB for displaying
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Convert the grayscale image to a NumPy array
    image_matrix = np.array(resized_image)

    # Print the matrix
    print(image_matrix)

    # Display the images
    plt.figure(figsize=(10, 5))

    # Display the original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    # Display the original image - gray version
    plt.subplot(1, 3, 2)
    plt.imshow(gray_image)
    plt.title('Original Image gray version')
    plt.axis('off')

    # Display the resized image
    plt.subplot(1, 3, 3)
    plt.imshow(resized_image, cmap='gray')
    plt.title('Resized 30x30 Image')
    plt.axis('off')

    plt.show()
else:
    print("Please upload exactly one image.")
```
