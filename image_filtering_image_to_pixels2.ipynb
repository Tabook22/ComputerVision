```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload files
uploaded = files.upload()
```

```python
# Ensure only one image is uploaded
if len(uploaded) == 1:
    filename = next(iter(uploaded))
    # Read the image using OpenCV
    original_image = cv2.imread(filename)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
    # Resize the image to 12x12 pixels
    resized_image = cv2.resize(gray_image, (30, 30))

    # Convert the original BGR image to RGB for displaying
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Convert the grayscale image to a NumPy array
    image_matrix = np.array(resized_image)

    # Print the matrix
    print(image_matrix)

    # Display the images
    plt.figure(figsize=(10, 5))

    # Display the original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    # Display the original image - gray version
    plt.subplot(1, 3, 2)
    plt.imshow(gray_image, cmap='gray')
    plt.title('Original Image gray version')
    plt.axis('off')

    # Display the resized image
    plt.subplot(1, 3, 3)
    plt.imshow(resized_image, cmap='gray')
    plt.title('Resized 30x30 Image')
    plt.axis('off')

    plt.show()
else:
    print("Please upload exactly one image.")
```

Here we are going to use the Kernel or the filter, to blure the Images (Box Blure).

Keep in mind that the **cv2.filter2D()** function to apply a box blur to the grayscale version of the input image. The box blur is a type of blurring that averages the colors of a rectangle of pixels around each pixel in the image. The larger the rectangle, the more blurred the image will be (3x3,5x5,7x7).

The cv2.filter2D() function takes three arguments:

The image to be filtered<br>
The depth of the image (which is usually -1 for grayscale images)<br>
The kernel to be applied<br>
<p>
The kernel is a small matrix of numbers that defines the weights of the pixels in the rectangle around each pixel. In this case, the kernel is a 3x3 matrix of all ones, which means that all pixels in the rectangle will have equal weight.
</p>
<p>
To increase the blurring, you can increase the size of the rectangle. This will make the averaging process more effective, resulting in a more blurred image. </p>

```python
# Upload files
uploaded = files.upload()
# Ensure only one image is uploaded
if len(uploaded) == 1:
    filename = next(iter(uploaded))
    # Read the image using OpenCV
    original_image = cv2.imread(filename)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

    # Define the 5x5 box blur kernel
    #kernel = np.ones((5, 5), dtype=np.float32) / 25

    # Define the 3x3 box blur kernel
    kernel = np.ones((3, 3), np.float32) / 9
    #Dividing by 9 in the box blur kernel is crucial for ensuring that the filter performs an averaging operation.

    # Apply the kernel (box blur) to the image, the cv2.filter2D() function to apply a box blur to the grayscale version of the input image
    # The -1 means the image is a gray scale
    blurred_image = cv2.filter2D(gray_image, -1, kernel)

    #Here i can increase the amount of blurring by applying the blurring more that one time
    for i in range(20):
      blurred_image = cv2.filter2D(blurred_image, -1, kernel)

    # Convert the original BGR image to RGB for displaying
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # Display the images
    plt.figure(figsize=(10, 5))

    # Display the original image
    plt.subplot(1, 2, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    # Display the blurred image
    plt.subplot(1, 2, 2)
    plt.imshow(blurred_image, cmap='gray')
    plt.title('Box Blurred Image')
    plt.axis('off')

    plt.show()
else:
    print("Please upload exactly one image.")
```

How we can increase the blurring effect?

increase the blurring effect in image processing, you can modify the box blur kernel or use alternative blurring techniques. Here are some methods to achieve a stronger blurring effect:

1. Use a Larger Kernel:

Increasing the size of the kernel will increase the blurring effect. Instead of a 3x3 kernel, you could use a 5x5, 7x7, or even larger kernel. Remember to normalize the kernel so that its elements sum up to 1. For instance, in a 5x5 kernel, each element would be
1/25
 .
2. Apply the Blur Multiple Times:

Repeatedly applying the same blurring filter can increase the blurring effect. This is equivalent to convolving with a larger, more spread-out kernel.

3. Gaussian Blur:
Instead of a box blur, use a Gaussian blur. Gaussian blur uses a kernel where the values are sampled from a Gaussian function. It typically produces a more natural and visually pleasing blur than a box blur.
Median Blur:

4. Median blur replaces each pixel's value with the median value of the intensities in the kernel neighborhood. It's useful for removing noise from an image while keeping the edges relatively sharp.

5. Bilateral Filter:
Bilateral filtering is a technique that blurs an image while preserving edges. It does this by taking into account the difference in intensity between neighboring pixels.
