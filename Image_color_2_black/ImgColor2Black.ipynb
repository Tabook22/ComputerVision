```python
import cv2
import matplotlib.pyplot as plt

# Load the colored image
image_path = 'abc.jpg'  # replace with the path to your image
colored_image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)

# Save the grayscale image
cv2.imwrite('gray_image.jpg', gray_image)

# Display the original and grayscale images side by side
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```
