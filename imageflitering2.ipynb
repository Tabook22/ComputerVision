```python
from google.colab import files
from google.colab.patches import cv2_imshow

import cv2
import numpy as np

# Upload the image
uploaded = files.upload()

# Read the image
image = cv2.imread(list(uploaded.keys())[0])

# Convert the image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Blurring kernel
blurring_kernel = np.ones((3, 3), dtype=np.float32) / 9

# Sharpening kernel
sharpening_kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

# Edge detection kernel
edge_detection_kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

# Embossing kernel
embossing_kernel = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])

# Apply the blurring kernel
blurred_image = cv2.filter2D(image_gray, -1, blurring_kernel)

# Apply the sharpening kernel
sharpened_image = cv2.filter2D(image_gray, -1, sharpening_kernel)

# Apply the edge detection kernel
edge_detected_image = cv2.filter2D(image_gray, -1, edge_detection_kernel)

# Apply the embossing kernel
embossed_image = cv2.filter2D(image_gray, -1, embossing_kernel)

# Display the original image
cv2_imshow(image)

# Display the blurred image
cv2_imshow(blurred_image)

# Display the sharpened image
cv2_imshow(sharpened_image)

# Display the edge-detected image
cv2_imshow(edge_detected_image)

# Display the embossed image
cv2_imshow(embossed_image)

# Wait for a key press before closing the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
